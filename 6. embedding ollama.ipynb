```python
import ollama

f = open("data/embeddings/keys.txt", "r")
keys = f.read().split()
f.close()
```

```python
import pandas as pd

# df, each row is a key
df = pd.DataFrame(columns=['key', 'embedding'])
df['key'] = keys

for i, row in df.iterrows():
    print(i, row['key'])
    e = ollama.embed(
        model='mxbai-embed-large',
        input=row['key']
    )
    df.at[i, 'embedding'] = e.embeddings
```

```python
df.to_csv("data/embeddings/small.csv", index=False)
```
